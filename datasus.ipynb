```
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```
# aquisição de dados do sus
dados = pd.read_csv("/content/A190839189_28_143_208.csv",encoding="ISO-8859-1",skiprows=3,sep=";",skipfooter=12,thousands=".",decimal=",",na_values="-")
```

```
pd.options.display.float_format="{: .2f}".format
dados
```

```
dados['2008/Jan'].mean()
```

```
dados.plot(x="Unidade da Federação", y="2008/Ago",kind="bar", figsize=(10,7))
```

```
ax = dados.plot(x="Unidade da Federação", y="2008/Ago",kind="bar", figsize=(10,7))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.title("Gastos de Procedimento Hospitalar do SUS por Unidade da Federação")
```

```
dados['2020/Jan'].mean()
```

```
# Plotar gráfico mais recente

last_month = dados.plot(x="Unidade da Federação", y = dados.columns[-2],kind="bar", figsize=(10,7))
last_month.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.title("Gastos de Procedimento Hospitalar do SUS por Unidade da Federação")
```
