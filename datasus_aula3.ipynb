## Desafio 01: Escolher uma palete de cores mais adequada do matplotlib.





```
import matplotlib.pyplot as plt
import matplotlib

def new_palette():
  cmap = getattr(matplotlib.cm,  'rainbow')
  cmap.colorbar_extend
  colors = []
  norm = matplotlib.colors.Normalize(vmin=0, vmax=30)
  aux = [cmap(norm(value), alpha = 0.75) for value in range(0,30)]              
  # arredondar cada codigo RGBA com 3 casas decimais
  for item in aux:
    colors.append([ round(v,3) for v in item ])
  return colors

ax = ordenado_por_total[:].T.plot(color=new_palette())
#plt.imshow([colors])
plt.legend(loc='best', bbox_to_anchor=(1.25, 1.0),
          ncol=2, fancybox=True, shadow=True)
```
